Fitting a model of the form

$$
f(t) = \frac{K}{1+e^{a + \lambda t}}
$$

$K$ is known.

Data: $(t_1, y_1), (t_2, y_2),\ldots,(t_n,y_n)$

Logit function:

$$
l(u) = \log(\frac{u}{1-u})
$$

Apply logit to the $y$ data.... (logit transform the data0

New Data: $(t_1, l(y_1)), (t_2, l(y_2)),\ldots,(t_n,l(y_n))$

Then you have a LINEAR model:

$$
l(y_i) = a + \lambda t_i
$$

Then do least squares/linear regression and get estimates for $a$ and $\lambda$ with p-values from regression and all that.
